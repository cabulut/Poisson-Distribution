```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from scipy.stats import poisson
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

**Web Scraping**

```python
def table(league='premier-league-tabelle',teams=20):
    
    page = 'https://sport.sky.de/' + league
    request = requests.get(page)
    soup = BeautifulSoup(request.text, 'html.parser')
    table = soup.find("table",{"class":"sdc-site-table"})
    
    games = []
    for team in table.find_all("tbody"):
        rows = team.find_all("tr")
        for i in range(1,9):
            for row in rows:
                lst = row.find_all("td",{"class":"sdc-site-table__cell"})[i].text.strip()
                games.append(lst)
        data = [games[x:x+teams] for x in range(0, len(games), teams)]
        df = pd.DataFrame.from_records(data).transpose()
        df = pd.DataFrame(data).transpose()
        df.rename(columns={0:'team',1:'match',2:'win',3:'draw',4:'lose',5:'goal',6:'gd',7:'pts'},inplace=True)
        df = pd.concat([df[['team','match','win','draw','lose','gd','pts']], df['goal'].str.split(':', expand=True)], axis=1)
        df.rename(columns={0:'goal',1:'conceded'},inplace=True)
        df = df[['team','match','win','draw','lose','goal','conceded','gd','pts']]
        df['gd'] = df['gd'].str.replace('+','').str.replace('-','')
        df[df.columns[1:]] = df[df.columns[1:]].astype('int')
        return df
```

**Applying Function to Key and Value Pairs of Leagues**

```python
leagues = {"bundesliga-tabelle":18,"premier-league-tabelle":20,"serie-a-tabelle":20,"ligue-1-tabelle":20,"la-liga-tabelle":20,"eredivisie-tabelle":18,"sueper-lig-tabelle":21}

#leagues and number of teams in these leagues
```

```python
bl,pl,seriea,ligue1,laliga,dutch,tsl = [table(league=k,teams=v) for k,v in leagues.items()]

#apply the function to get league table of leagues
```

**Poisson Distribution Function**

```python
val = range(10)
def goal_pmf(df=pl):
    global val
    if len(df) % 2 == 0:
        goal = df['goal'].sum() / (df['match'].max() * (len(df) / 2) - ((df['match'].max() * len(df)) -  df['match'].sum())/ 2)
        goal = poisson(goal)
        pmf = goal.pmf(val)
        return pmf
    else:
        goal = df['goal'].sum() / (df['match'].max() * ((len(df) - 1) / 2) - ((df['match'].max() * len(df)) -  df['match'].sum()) / 2)
        goal = poisson(goal)
        pmf = goal.pmf(val)
        return pmf
```

```python
bundesliga,super_lig,serie_a,ligue_1,la_liga,eredivisie,premier_league = (df.pipe(goal_pmf) for df in [bl,tsl,seriea,ligue1,laliga,dutch,pl])

#applying Poisson Distribution to Leagues
```

**Plotting**

```python
index = ["bundesliga","super lig","serie a","ligue 1","la liga","eredivisie","premier league"]
df = pd.DataFrame([bundesliga,super_lig,serie_a,ligue_1,la_liga,eredivisie,premier_league],index=index).transpose()
sns.set()
df.plot(figsize=(10,5),linewidth=3)
plt.title('Total Number of Goals Possibility',fontsize=15)
plt.xlabel('Total Goal',fontsize=15)
plt.ylabel('Possibility of Goal',fontsize=15)
plt.tight_layout()
plt.show()
```
